```python
nlist = []
for i in tlist:
    nlist.append(i.strip())
nlist
```

```python
#Dictionary creation
numL = []
namL = []
champDic = {}

for line in tlist:
    numL.append(line.split(":")[0])

for line in tlist:
    namL.append(line.split(":")[1].strip("\n"))

for key in numL:
    for value in namL:
        champDic[key] = value
        namL.remove(value)
        break

#save file in pickle
pikFile = open("C:/Users/Aaron/Desktop/loldfs/champDic.pkl", "wb")
pickle.dump(champDic, pikFile)
pikFile.close()

```
