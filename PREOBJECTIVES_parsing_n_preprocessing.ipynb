```python
with open("champion.txt", "r", encoding = "utf-8") as f:
    tlist = f.readlines()
print(tlist)
```

```python
nlist = []
for i in tlist:
    nlist.append(i.strip())
nlist
```

```python
numL = []
namL = []
champDic = {}

for line in tlist:
    numL.append(line.split(": ")[0])

for line in tlist:
    namL.append(line.split(": ")[1].strip("\n"))

for key in numL:
    for value in namL:
        champDic[key] = value
        namL.remove(value)
        break

champDic
```

```python
pikFile = open("C:/Users/Aaron/Desktop/loldfs/OBJECTIVE_ONE_champion_ID.pkl", "wb")
pickle.dump(champDic, pikFile)
pikFile.close()
```
