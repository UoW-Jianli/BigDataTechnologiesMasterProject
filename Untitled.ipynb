```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import numpy as np
import json
import re 
import time
import seaborn as sns
```

```python
match = pd.read_csv("C:/Users/Aaron/Desktop/loldfs/objective2_encoded.csv")
```

```python
matchWin = match[match["win"] == 1]
matchLose = match[match["win"] == 0]
```

```python
winCount = pd.concat([matchWin["p1"], matchWin["p2"], matchWin["p3"], matchWin["p4"], matchWin["p5"]], keys = ["x", "y", "z", "w", "v"])
winCount = pd.DataFrame(winCount, columns = ["counts"])
winCount
```

```python
totWin = pd.DataFrame(winCount["counts"].value_counts(), columns = ["counts"])
totWin = totWin.rename(columns = {"counts": "wins"})
totWin
```

```python
loseCount = pd.concat([matchLose["p1"], matchLose["p2"], matchLose["p3"], matchLose["p4"], matchLose["p5"]], keys = ["x", "y", "z", "w", "v"])
loseCount = pd.DataFrame(loseCount, columns = ["counts"])
loseCount
```

```python
totLose = pd.DataFrame(loseCount["counts"].value_counts(), columns = ["counts"])
totLose = totLose.rename(columns = {"counts": "lose"})
totLose
```

```python
wl = pd.concat([totWin, totLose], axis = 1, join = "inner")
wl
```

```python
winrate = pd.DataFrame(wl["wins"] / (wl["wins"] + wl["lose"]) * 100, columns = ["winrate"])
winrate
```

```python
winrate = winrate.sort_values(by = "winrate", ascending = False)
winrate
```

```python
winrate = winrate.reset_index()
winrate
```

```python
winrate = winrate.rename(columns = {"index": "champion"})
winrate
```

```python
fig_dims = (15, 45)
fig, ax = plt.subplots(figsize = fig_dims)
sns.barplot(x = "winrate", y = "champion", data = winrate, ax = ax)
ax.set_title("Win rate champions Season 10 2020 (high elo), over 92k matches")
```

```python
wl = wl.reset_index()
wl = wl.rename(columns = {"index": "champion"})
wl
```

```python
fulldf = pd.merge(wl, winrate)
fulldf
```

```python
picks = fulldf["wins"] + fulldf["lose"]
picks
```

```python
fulldf = pd.concat([fulldf, picks], axis = 1)
fulldf
```

```python
fulldf = fulldf.rename(columns = {0: "totpick"})
fulldf
```

```python
fig_dims = (15, 45)
fig, ax = plt.subplots(figsize = fig_dims)
sns.barplot(x = "totpick", y = "champion", data = fulldf.sort_values(by="totpick", ascending = False), ax = ax)
ax.set_title("Tot picks champions")
```
